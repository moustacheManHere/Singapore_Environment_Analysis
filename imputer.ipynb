```python
# iterative is the best for categorical and numerical vars, knn is good for numerical, interpolate just using linear relation to do it
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer # doesnt work when i dont inclide te prev line
```

```python
# Create a sample pandas dataframe with missing values
df = pd.read_csv("./wrangled/manufacturing missing.csv")
# Ordinal encode the Year feature
df['year'] = df['year'].astype('category')
df['year_encoded'] = df['year'].cat.codes
# its called one hot encoding. bascially you make it easy for algorithm to deal with the data
# columns are indicated with 1 or 0 to show if categor applies there.
df = pd.get_dummies(df, columns=['level_2'], prefix=['type'])

df
```

```python

# I ma imputing missing values now
# i nvr do random_state cuz i dont mind the data changing
# iset max to be 10 as i think it is accurate enough and i dont want my prog to run too long
imputer = IterativeImputer(max_iter=10)
df_imputed = pd.DataFrame(imputer.fit_transform(df.drop(['year'], axis=1)), 
                          columns=df.drop(['year'], axis=1).columns)

```

```python
cols = df_imputed.iloc[:,2:]
# get only the catgory columns
newDf = pd.DataFrame(columns=["year","type","value"])
newDf["value"] = df_imputed["value"]
newDf["year"] = df["year"]
# bascially returns the max column for each row
newDf["type"] = cols.idxmax(axis=1)
newDf["type"] = newDf["type"].str.replace('type_', '')
newDf.to_csv("./wrangled/manufacturing imputed.csv",index=False)
```
